```julia
""" Simulate growth rate with  #docstring!rate 'lambda', initial value 'x0' and time 'N' """

function growth(lambda, x0, N) #creating a function
    x = zeros(N+1) #zeros function creates an array with all 0s with the same type and shape as N+1
    x[1] = x0 #julia indexing starts at 1
    
    for n in 1:N
        x[n+1] = lambda * x[n]
    end
    
    return x

end

    
    
```

```julia
#testing out the growth function
growth(1.2, 10.0, 10)
```

```julia
""" Generates growth and plots with rate 'lambda', initial value 'x0' and time 'N' """

function plot_growth(lambda, x0, N) #creating a function
    x = zeros(N+1) #zeros function creates an array with all 0s with the same type and shape as N+1
    x[1] = x0 #julia indexing starts at 1
    
    for n in 1:N
        x[n+1] = lambda * x[n]
    end

    return scatter(x)

end
```

```julia
using Plots
plot_growth(1.2, 10.0, 10)
```

# Solving ODEs

```julia
f(u,p,t) = 0.98u
```

```julia
using Pkg
Pkg.add("DifferentialEquations")
```

```julia
using DifferentialEquations
f(u,p,t) = 0.98u
u0 = 1.0
tspan = (0.0,1.0) # specfies the time interval of which the ODE should be solved
prob = ODEProblem(f,u0,tspan) # defines the initial value problem
```

```julia
sol = solve(prob)
```

```julia
using Plots; gr()
plot(sol)
```

```julia
plot(sol, linewidth=5,
     title="Solution to the linear ODE with a thick line",
     xaxis="Time (t)", yaxis="u(t) (in μm)",
     label="My Thick Line!") # legend=false
```

```julia
sol(0.45)
```

```julia
sol = solve(prob, saveat=0.1) # to save at specific time points, every 0.1 t
```

# Modelling Lotka Volterra System

```julia
""" Simulate predator and prey populations using LV model """
function lv(du,u,p,t)
    alpha, beta, delta, lambda = p
    du[1] = alpha*u[1] - beta*u[1]*u[2]
    du[2] = delta*u[1]*u[2] - lambda*u[2]
end
```

```julia
u0 = [30.0, 30.0]
```

```julia
p = (0.8, 0.3, 0.1, 1.2)
```

```julia
tspan = (0.0, 100.0)
prob = ODEProblem(lv, u0, tspan, p)
```

```julia
sol = solve(prob)
```

```julia
plot(sol, ylimits = (0, 100)) #yay
```
